## Extract daily price time series of the 10 largest Cryptocurrencies according to the most recent Market Cap from Coinmarketcap plus Bitcoin. For each series (11 series), go as far as possible in the past but discard initial periods where the prices are extremely small.

```python
def currency(url):
  content = requests.get(url).content
  soup = BeautifulSoup(content,'html.parser')
  table = soup.find('table', {'class': 'table'})
  data = [[td.text.strip() for td in tr.findChildren('td')] 
        for tr in table.findChildren('tr')]
  
  df = pd.DataFrame(data)
  df.drop(df.index[0], inplace=True) # first row is empty
  df[0] =  pd.to_datetime(df[0]) # date
  for i in range(1,7):
    df[i] = pd.to_numeric(df[i].str.replace(",","").str.replace("-","")) # some vol is missing and has -
  df.columns = ['Date','Open','High','Low','Close','Volume','Market Cap']
  df.sort_values(by=['Date'])
  df1=df[['Date','Close','Volume','Market Cap']] 
  df1=df1.iloc[:int(0.9*len(df1)),:]# discard initial periods where the prices are extremely small
  df1=df1.dropna()
  
  return df1
```

## Transform the series in percentage log returns, make separate plots, and make one descriptive statistics table reporting mean, standard deviation, skewness, kurtosis, autocorrelation coefficient of order 1. Comment and compare with your experience from stock returns in the previous homeworks.

```python
def logrtn(df):    
    df['Close']=np.log(df['Close'])
    Returns=np.diff(df['Close'])*100
    Returns=np.append(0,Returns)
    df['Returns']=Returns
    return df
```

```python
d=b
for i in range(len(d)):
    d[i]=d[i].iloc[::-1]
```

```python
for i in range(len(d)):
    d[i]=logrtn(d[i])
```

```python
for i in range(len(d)):
    plt.plot(d[i]['Date'],d[i]['Returns'],label =a[i])
    plt.ylabel('Log Returns')
    plt.xlabel('Date')
    plt.title(a[i]+' log returns')
    plt.show()
```

```python
stat_table=pd.DataFrame(np.arange(55).reshape((11,5)),columns=['mean', 'standard deviation', 'skewness', 'kurtosis', 'autocorrelation coefficient of order 1'
                                                             ],index=a)
```

```python
def auto_corrcoef(x):
    return np.corrcoef(x[1:-1], x[2:])[0,1]
```

```python
def table_base2(df):
    ls=[]
    ls.append(np.mean(df['Returns']))
    ls.append(np.std(df['Returns']))
    ls.append(stats.skew(df['Returns']))
    ls.append(stats.kurtosis(df['Returns']))
    ls.append(auto_corrcoef(df['Returns']))
    return ls
```

```python
for i in range(len(d)):
    stat_table.iloc[i,:]=table_base2(d[i])
```

Most of currencies' returns' skewness is close to zero, which means these returns' distribution is close to symmetrical. All the returns' kurtosis is positive, which means their distribution's tails are falttening. Finally, we can observe from the  autocorrelation coefficient of order 1 that almost all the returns' series can not use AR1 model to represent, which I can also observe from the graphs I did in the previous.

## For each of the Cryptocurrencies, estimate a GARCH (p,q) type model. Report in a table all the parameter estimates. Make individual plots of the estimated daily volatilities. Explain why you choose these p and q.

```python
sum_table=pd.DataFrame(np.zeros((11, 4)),columns=['mu', 'omega', 'alpha[1]', 'beta[1]'],index=a)
#Garch model with normal innovation density
for i in range(len(d)):
  c=d[i]['Returns']
  garchmodel_gaussian = arch_model(c, mean = 'Constant', vol='Garch', p=1, o=0, q=1,dist='normal')
  garchmodel_gaussian_res = garchmodel_gaussian.fit(update_freq=20, disp='off')
  print(garchmodel_gaussian_res.summary())
  X=range(0,len(garchmodel_gaussian_res.conditional_volatility))
  plt.figure(figsize=(10,6))
  plt.plot(X, garchmodel_gaussian_res.conditional_volatility, color='blue')
  plt.title(a[i]+' estimated daily volatilities')
  plt.show()
  results_summary = garchmodel_gaussian_res.summary()
  # Note that tables is a list. The table at index 1 is the "core" table. Additionally, read_html puts dfs in a list, so we want index 0
  results_as_html = results_summary.tables[1].as_html()
  results_as_html1 = results_summary.tables[2].as_html()
  a1=pd.read_html(results_as_html, header=0, index_col=0)[0]
  a2=pd.read_html(results_as_html1, header=0, index_col=0)[0]
  sum_table['mu'][i]=a1['coef'][0]
  sum_table['omega'][i]=a2['coef'][0]
  sum_table['alpha[1]'][i]=a2['coef'][1]
  sum_table['beta[1]'][i]=a2['coef'][2]
```

> We are staying with a GARCH(1,1) mode because it is the most commonly available, the most commonly used, and sometimes good enough. For the garch(1,1) model the key statistic is the sum of the two main parameters (alpha1 and beta1, in the notation we are using here). The sum of alpha1 and beta1 should be less than 1
